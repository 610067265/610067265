```python
import networkx as nx
%matplotlib notebook
import pandas as pd   

import matplotlib.pyplot as plt 
from community import community_louvain
import io
```

```python
# Reading the edgelist csv file 
df_edges = pd.read_csv('musae_ENGB_edges.csv')
df_edges.head()
```

```python
# Reading the node csv file
df_nodes = pd.read_csv("musae_ENGB_target.csv")
df_nodes.head()
```

```python
# Loading the graph with node attributes 
G = nx.from_pandas_edgelist(df_edges, 'from', 'to')
data = df_nodes.set_index('new_id').to_dict('index')
# Adding node attributes from the nodes data file 
nx.set_node_attributes(G, data)
```

```python
print(G.nodes(data=True))
```

```python
count_True = 0
for w in G.nodes:
  if G.nodes[w]["mature"]:
    count_True += 1
print(f"Number of True: {count_True} Number of Nodes: {len(G.nodes())} Ratio: {count_True/len(G.nodes())}")
```
